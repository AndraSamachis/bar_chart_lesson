```python
import pandas as pd
import matplotlib.pyplot as plt

file = 'dataset.csv'
df = pd.read_csv(file)

platforms = ['PS4', 'XOne', 'PC', 'WiiU']
df_filtered_by_platform = df[df['platform'].isin(platforms)]

df_genre_platform = df_filtered_by_platform[['platform', 'genre']]

genres = [
    'Action', 'Adventure', 'Fighting', 'Misc', 'Platform', 'Puzzle', 'Racing', 'Role-Playing',
    'Shooter', 'Simulation', 'Sports', 'Strategy'
]

genre_counts = {}
for platform in platforms:
    genre_counts[platform] = {}
    for genre in genres:
        genre_counts[platform][genre] = 0

for platform in platforms:
    for genre in genres:
        genre_counts[platform][genre] = df_genre_platform[
            (df_genre_platform['platform'] == platform) &
            (df_genre_platform['genre'] == genre)
        ].shape[0]

counts_df = pd.DataFrame(genre_counts).T

plt.figure(figsize=(12, 6))
counts_df.plot(kind='bar')
plt.xlabel('platform')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.legend(title='genre', loc = 'upper left', bbox_to_anchor=(1, 1.02))
plt.grid(axis='x')

plt.tight_layout()
plt.show()
```
